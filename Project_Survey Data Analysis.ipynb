#Using series.string.split to split each row in the series based a delimited and passing to that #function as an argument. In this case, the delimiter is ';' and used the.str.split(';').
#Using argument expand=True, to str.split(). This will create a new dataframe from our series by making each language its own column (each row still represents one respondent).

```python
#Breaking the response'LanguageHaveWorkedWith' to drill down and bring up clear picture by deeper analysis 
print('Insights: Each row in the below series is now a row in the new data frame, and each language has been split from the others into unique columns.')
lang_lists = df_survey['LanguageHaveWorkedWith'].str.split(';', expand=True)
lang_lists.head(10)
```

#Finding out the number of times each language was mentioned in total:
#Using df.stack() to stack this DataFrame, slicing each column and then stacking them on top of each other so that every data point in the DataFrame appears in a single pandas Series.
#Using value_counts() on this new “stacked” series to get the total number of times each language is mentioned.

```python
print('Every answer stacked into a single series as given below:')
lang_lists.stack().head(20)
```

```python
print('Every answer stacked into a single series as given below:')
lang_lists.stack().value_counts()
```

```python
#Potting Bar Graph for LanguageWorkedwith Response
lang_lists.stack().value_counts().plot(kind='bar', figsize=(15,7), color="#61d199")
```
